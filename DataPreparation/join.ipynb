```python
import pandas as pd
```

```python
# Creating dataframe a
a = pd.DataFrame()
 
# Creating Dictionary
d = {'id': [1, 2, 10, 12], 
     'val1': ['a', 'b', 'c', 'd']}
 
a = pd.DataFrame(d)
 
# printing the dataframe
a
```

```python
# Creating dataframe b
b = pd.DataFrame()
 
# Creating dictionary
d = {'id': [1, 2, 9, 8],
     'val1': ['p', 'q', 'r', 's']}
b = pd.DataFrame(d)
 
# printing the dataframe
b
```

## Pandas Inner Join
Pandas bietet die Funktion merge

### Argumente 
a und b: die beiden Tabellen die gejoint werden sollen

on: Spalte für die beide Tabellen verglichen werden, analog der SQL Syntax

how: Welche Art von Join, z.b. 'inner' für Inner Join


SELECT *
FROM df1
INNER JOIN df2
  ON df1.key = df2.key;


```python
# inner join
# Schnittmenge der beiden Tabellen im Vergleich mit dem Primär- und Fremdschlüssel
# Hier ist der Schlüssel 'id'
# Nur zwei Zeilen haben den selben Schlüssel und können gejoint bzw. gemerged werden
df1 = pd.merge(a, b, on='id', how='inner')
 
# display dataframe
df1
```

```python
# left outer join
# Das Ergebnis ist die Menge aus Tabelle a und die Schnittmenge
# Für die id 10 und 12 fehlt val1_y, daher sind hier die Einträge NaN
df2 = pd.merge(a, b, on='id', how='left')
 
# display dataframe
df2
```

```python
# right outer join
# Das Ergebnis ist die Menge aus Tabelle b und die Schnittmenge
# Die fehlenden Werte für val1_x mit dem Index 2 und 3 sind NaN
df3 = pd.merge(a, b, on='id', how='right')
 
# display dataframe
df3
```

```python
# full outer join
# Entspriche der Vereinigung der Mengen
# Zu id 10 und 12 gibt es keine Einträge für val1_y, zur id 8 und 9 gibt es keine Einträge für val1_x
df4 = pd.merge(a, b, on='id', how='outer')
 
# display dataframe
df4
```
