```python
import pandas as pd
%time
tps_october = pd.read_csv("../../train_0.csv")
```

```python
tps_october.head()
```

```python
def big_function(col1, col2, col3):
    return (col1 ** 10 + col2 ** 9 + np.sqrt(col3 ** 3))
```

```python
%time tps_october['f1000'] = tps_october.apply(lambda row: big_function(row['ball_pos_x'], row['ball_pos_y'], row['ball_pos_z']), axis=1)
```

```python
%time 
tps_october['f1001'] = big_function(tps_october['ball_pos_x'].values, 
                                          tps_october['ball_pos_y'].values, 
                                          tps_october['ball_pos_z'].values)
```

# Looping vs. Vectorization
Anwenden der Funktion big_function mit apply(lambda: ...)

Dauer 11.2s

Mit Vektorisierung

# Lambda Functions
Für kleinere Funktionen muss die Funktion nicht explizit definiert werden, sondern es kann eine Lambda Fun

```python
words = ["apple", "banana", "cherry", "date"]
sorted_words = sorted(words, key=lambda x: len(x))
print(sorted_words)
```
